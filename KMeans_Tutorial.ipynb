```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=400, n_features=2, centers=5, cluster_std=0.5, shuffle=True, random_state=4)
```

##Plotting the dataset

```python
plt.figure(figsize=(12, 10))
plt.scatter(
   X[:, 0], X[:, 1],
   c='red', marker='o',
   edgecolor='black', s=40
)
plt.show()

colors = 10*['blue','red','pink','green','yellow','cyan','magenta']
```

## Plotting the K Means clusters for varying values of K

```python
from sklearn.cluster import KMeans
from matplotlib.pyplot import figure
```

##**K-Means for k=3**

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
kmeans.labels_
```

```python
plt.figure(figsize=(12, 10))
for i in range(len(X)):
  plt.scatter(
   X[i][0], X[i][1],
   c=colors[kmeans.labels_[i]], marker='o',
   edgecolor='black', s=40)
  
for i in range(len(kmeans.cluster_centers_)):
  plt.scatter(
   kmeans.cluster_centers_[i][0], kmeans.cluster_centers_[i][1],
   c='black', marker='*',
   edgecolor='black', s=60,linewidths=4)
  
figure(figsize=(10, 8), dpi=100)
```
