```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=0.5, shuffle=True, random_state=4)
```

```python
plt.scatter(
   X[:, 0], X[:, 1],
   c='red', marker='o',
   edgecolor='black', s=40
)
plt.show()
```
