#K-Means Clustering Tutorial

This notebook is used for explaining the steps involved in creating a K-Means Clustering model 

1. Import the required libraries
2. Generate dataset
3. Plotting the dataset
4. Training and plotting the K Means clusters for varying values of K


## Import the required libraries

```python
from sklearn.cluster import KMeans        #Used to create KMeans Model
from sklearn.datasets import make_blobs   #Used to generate dataset
import matplotlib.pyplot as plt           #Used to for making plots 
```

## Generate Dataset

Data set is generated with 400 data points and 2 features having 5 centers

```python
X, y = make_blobs(n_samples=400, n_features=2, centers=5, cluster_std=0.5, shuffle=True, random_state=4)      
```

##Plotting the dataset

```python
X1 = X[:, 0]                 # Storing the list of 1st feature
X2 = X[:, 1]                 # Storing the list of 2nd feature
plt.scatter(X1, X2, c='blue', marker='o', edgecolor='black', s=35)   # Plotting the data points
plt.show()
```

```python
map_label_to_color = 10*['blue','red','pink','green','yellow','cyan','magenta']     #Map of cluster label to colors to color the clusters
```

## Training and plotting the K Means clusters for varying values of K

###**K-Means for k=3**

```python
model_kmeans = KMeans(n_clusters=3, random_state=0)           # Creating K Means model with 3 clusters
model_kmeans = model_kmeans.fit(X)                            # Fitting the K Means model
```

```python
model_kmeans.labels_                                          # Printing the labels of the clusters each datapoint belongs to
```

```python
for i in range(len(X)):                                          # Iterating over each data point 
  x1 = X[i][0]
  x2 = X[i][1]
  plt.scatter( x1, x2, c=map_label_to_color[model_kmeans.labels_[i]], marker='o', edgecolor='black', s=40)    #Plotting the data point

cluster_center_list = model_kmeans.cluster_centers_                       # Storing the list of centers
for i in range(len(cluster_center_list)):                                 # Iterating over the centers
  plt.scatter( cluster_center_list[i][0], cluster_center_list[i][1], c='black', marker='*', edgecolor='black', s=60,linewidths=4) #Printing the cluster centers
```
